```python
def opcode(input_data):
    output = input_data
    #print(output)
    
    k = 0
    new_data = input_data
    while len(new_data) >= 2:
        if new_data[0] == 99:
            break;
        else:
            #read the first instruction:
            print('first instruction: ', new_data[0])
            ins = [int(x) for x in str(new_data[0])]
            #print('ins: ', ins)
            if len(ins) == 1:
                c = ins[0]
                ins = [0, 0, 0, 0, c]
            else:
                padding = [0]*(5-len(ins))
                ins = padding + ins
                c = 10*ins[-2] + ins[-1]
           
            if c == 1 or c == 2:
                k += 4
                
                if ins[2] == 0:
                    va1 = output[new_data[1]]
                else:
                    va1 = new_data[1]
                if ins[1] == 0:
                    va2 = output[new_data[2]]
                else:
                    va2 = new_data[2]
                
                idx3 = new_data[3]
                if c == 1:
                    output[idx3] = va1 + va2
                elif c == 2:
                    output[idx3] = va1 * va2
            
            elif c == 3 or c == 4:
                k += 2
                if ins[2] == 0:  #address mode
                    v = output[new_data[1]] 
                    idx1 = new_data[1]
                else:
                    v = new_data[1]
                    idx1 = 1
                #print(new_data)
                if c == 3:
                    inp_val = input()
                    output[idx1] = int(inp_val)
                elif c == 4:
                    print('code: ', v)
            elif c == 5:
                #jump if true:
                if ins[2] == 0:
                    param1 = output[new_data[1]]
                else:
                    param1 = new_data[1]
                if ins[1] == 0:
                    param2 = output[new_data[2]]
                else:
                    param2 = new_data[2]
        
                if param1: 
                    k = param2
                else:
                    k += 3
                
            elif c == 6:
                #jump if false:
                if ins[2] == 0:
                    param1 = output[new_data[1]]
                else:
                    param1 = new_data[1]
                if ins[1] == 0:
                    param2 = output[new_data[2]]
                else:
                    param2 = new_data[2]
        
                if not param1: 
                    k = param2
                else:
                    k += 3
            elif c == 7:
                k = k+4
                if ins[2] == 0:
                    v1 = output[new_data[1]]
                else:
                    v1 = new_data[1]
                if ins[1] == 0:
                    v2 = output[new_data[2]]
                else:
                    v2 = new_data[2]
                if v1 < v2:
                    output[new_data[3]] = 1
                else:
                    output[new_data[3]] = 0
            elif c == 8:
                k = k+4
                if ins[2] == 0:
                    v1 = output[new_data[1]]
                else:
                    v1 = new_data[1]
                if ins[1] == 0:
                    v2 = output[new_data[2]]
                else:
                    v2 = new_data[2]
                if v1 == v2:
                    output[new_data[3]] = 1
                else:
                    output[new_data[3]] = 0
            else:
                print('whaaaaat diiis??')
            
            new_data = output[k:]
           # print(output)
    del input_data
    del new_data
    return output
```
